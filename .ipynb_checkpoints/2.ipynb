```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Creating a DataFrame
df = pd.DataFrame({
    'StudentID': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Maths': [75, 82, 90, 65, 50],
    'Science': [80, 78, 92, 70, 45],
    'English': [70, 85, 88, 75, 60],
    'SocialStudies': [85, 75, 92, 60, 55],
    'Attendance': [90, 92, np.nan, 85, 80],
    'Gender': ['M', 'F', 'M', 'F', 'F']
})
```

```python
# 1. Dealing with missing values
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)
```

```python
# 2. Dealing with outliers
Q1 = df['Maths'].quantile(0.25)
Q3 = df['Maths'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR

df.loc[df['Maths'] > upper_limit, 'Maths'] = df['Maths'].median()
```

```python

# 3. Applying data transformation
# Here, we will transform the "Maths" column to a normal distribution using a logarithmic transformation.
df['Maths'] = np.log(df['Maths'])
```

```python
# Plotting the transformed "Maths" column
sns.histplot(df['Maths'])
```
